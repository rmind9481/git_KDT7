#### [ 자연어 - 한국어 형태소 분석기 활용 ]

[1] 모듈 로딩 <hr>

```python
## konlpy 내 모든 형태소 분석기 로딩
from konlpy.tag import * 

## konlpy 내 말뭉치 로딩
from konlpy.corpus import *
```

[2] 코퍼스/말뭉치 준비<hr>

```python
## 파일명 
LAW_FILE = kolaw.fileids()[0]

BILL_FILE_LIST = kobill.fileids()

print(f'LAW_FILE       => {LAW_FILE}')
print(f'BILL_FILE_LIST => {len(BILL_FILE_LIST)}개')
```

```python
## 말뭉치/코퍼스 로딩 
kolaw_corpus = kolaw.open(LAW_FILE).read()

print(f'길이 : {len(kolaw_corpus)}개')
print(f'내용 :\n{kolaw_corpus[:100]}')
```

```python
## 말뭉치/코퍼스 - 국회의안 
kobill_corpus=''
for filename in BILL_FILE_LIST: 
    kobill_corpus += kobill.open(filename).read()
    print(f'파일 : {filename} -------------------')
    print(f'길이 : {len(kobill_corpus)}개')
    print(f'내용 :\n{kobill_corpus[:100]}')

print(f'국회의안 말뭉치------------------------')
print(f'총길이 : {len(kobill_corpus)}개')
print(f'내용 :\n{kobill_corpus[:100]}')
```

[3] 다양한 형태소 분석 <hr>

```python
## 형태소 분석기 생성
han      = Hannanum()
kkma     = Kkma()
okt      = Okt()
komoran  = Komoran()
#mecab    = Mecab()  ## Windows OS 미지원! 
```

```python
## 명사 추출 메서드 = > 형태소분석기객체.nouns(데이터)
data = kolaw_corpus[:100]

## 형태소 분석기별 명사 추출
han_nouns = han.nouns(data)
kkma_nouns = kkma.nouns(data)
komoran_nouns = komoran.nouns(data)
okt_nouns = okt.nouns(data)

## 형태소 분석기별 결과 출력
print(f'Hannaum 분석기  =>\n{han_nouns}')
print(f'kkma 분석기     =>\n{kkma_nouns}')
print(f'Komoran 분석기  =>\n{komoran_nouns}')
print(f'Okt 분석기      =>\n{okt_nouns}')
```

```python
## 형태소 추출 메서드 = > 형태소분석기객체.morphs(데이터)
## 형태소 : 의미를 가지는 가장 작은 단위 

## 형태소 분석기별 형태소 추출
han_morphs      = han.morphs(data)
kkma_morphs     = kkma.morphs(data)
komoran_morphs  = komoran.morphs(data)
okt_morphs      = okt.morphs(data)

## 형태소 분석기별 결과 출력
print(f'Hannaum 분석기  =>\n{han_morphs}')
print(f'kkma 분석기     =>\n{kkma_morphs}')
print(f'Komoran 분석기  =>\n{komoran_morphs}')
print(f'Okt 분석기      =>\n{okt_morphs}')
```

```python
## 형태소+품사 메서드 = > 형태소분석기객체.pos(데이터)
## 형태소 : 의미를 가지는 가장 작은 단위 

## 형태소 분석기별 형태소 추출
han_pos      = han.pos(data)
kkma_pos     = kkma.pos(data)
komoran_pos  = komoran.pos(data)
okt_pos      = okt.pos(data)

## 형태소 분석기별 결과 출력
print(f'Hannaum 분석기  =>\n{han_pos}')
print(f'kkma 분석기     =>\n{kkma_pos}')
print(f'Komoran 분석기  =>\n{komoran_pos}')
print(f'Okt 분석기      =>\n{okt_pos}')
```

```python
## 형태소 분석기별 품사명 ==> 형태소분석기변수명.tagset
print(f'Hannaum 분석기  =>\n{han.tagset}')
print(f'Kkma    분석기  =>\n{kkma.tagset}')
print(f'Komoran 분석기  =>\n{komoran.tagset}')
print(f'Okt     분석기  =>\n{okt.tagset}')
```
