[3] 임의의 정수값으로 구성된 3차원 텐서를 아래 조건 만족하도록 생성하세요.

    * 생성 후 3개 2D 텐서로 분리 저장

```python
import numpy as np

tensor = np.random.rand(3, 10, 10)  # (3, 10, 10) 텐서 생성

# 첫 번째 차원 기준으로 나누기 (각각 (10, 10) 형태)
splits = [tensor[i, :, :] for i in range(tensor.shape[0])]  # 첫 번째 차원 기준으로 분리
print([split.shape for split in splits])  # 출력: [(10, 10), (10, 10), (10, 10)]
```

```python

t3 = torch.randint(0,256,(3,10,10) ,dtype=torch.uint8 )


# 2. 생성된 텐서를 첫 번째 차원 기준으로 2D 텐서 3개로 분리
split_tensors = [t3[i, :, :] for i in range(t3.shape[0])]

# 출력: 생성된 텐서와 분리된 2D 텐서들 확인
print("전체 3D 텐서 (3, 10, 10):")
print(tensor)

print("\n분리된 2D 텐서들:")
for idx, split_tensor in enumerate(split_tensors):
    print(f"2D 텐서 {idx+1} (10, 10):")
    print(split_tensor)
```

[5] 아래 2개 텐서에 대한 연산 수행 결과 출력하세요.
- t1 :  [[11,22], [33,44]]
- t2 :  [3,4,5]
- 연산 : 덧셈, 뺄셈, 곱셈, 나눗셈 연산자와 메서드

```python
t1 = torch.tensor([[11,22],[33,44]])
t2 = torch.tensor([3,5])

print(f'덧셈 : {t1+t2}')
print(f'뺄셈 : {t1-t2}')
print(f'나눗셈 : {t1/2}')
print(f'곱셈 : {t1*t2}')
print(f'행렬곱셈 : {t1@t2}')
```

[7] 3의 배수로 구성된 원소갯수가 12개인 텐서를 생성하세요.
- 1D 텐서 생성 T1
- T1의 축 추가로 2D 텐서 생성 T2
- T1의 축 추가로 3D 텐서 생성 T3
- T1의 축 추가로 4D 텐서 생성 T4

- T4의 축 축소로 3D 텐서 생성 T5
- T3의 축 축소로 2D 텐서 생성 T6
- T2의 축 축소로 1D 텐서 생성 T7

```python
t1 = torch.arange(3,37,3)
print(f'텐선 T1 ({t1}')

# T1의 축을 추가하여 2D 텐서 T2 생성
t2 = t1.unsqueeze(0)  # T1의 첫번째 축에 크기 1을 추가하여 2D 텐서로 변환
print(f'텐선 T2 ({t2}')


# T1의 축을 추가하여 3D 텐서 t3 생성
t3 = t1.unsqueeze(0).unsqueeze(0)
print(f'텐선 T3 ({t3}')

# T1의 축을 추가하여 4D 텐서  t4 생성
t4 = t1.unsqueeze(0).unsqueeze(0).unsqueeze(0)
print(f'텐선 T4 ({t4}')

# t4 의 축을 축소하여 3D 텐서 t5 생성
t5 = t4.squeeze(0)
print(f'텐선 T5 ({t5}')

# t3 의 축을 축소하여 2D 텐서 t6 생성

t6 = t3.squeeze(0)
print(f'텐선 T6 ({t6}')


# t2 의 축을 축소하여 1D 텐서 t7 생성
t7 = t2.squeeze(0)
print(f'텐선 T7 ({t7}')


```
