[3] 임의의 정수값으로 구성된 3차원 텐서를 아래 조건 만족하도록 생성하세요.
- 조건
    * 형태 : 10행 10열 3개
    * 타입 : 0 ~ 255값 저장되는 최소 크기 타입
    * 생성 후 3개 2D 텐서로 분리 저장

```python
# 0 ~ 255 사이의 임의의 정수값으로 3차원 텐서 생성 (형태: 10x10x3)
tensor_3d = torch.randint(0, 256, (10, 10, 3), dtype=torch.uint8)

# 3차원 텐서에서 각 2D 텐서를 분리
tensor_2d_1 = tensor_3d[:, :, 0]  # 첫 번째 채널 (10x10)
tensor_2d_2 = tensor_3d[:, :, 1]  # 두 번째 채널 (10x10)
tensor_2d_3 = tensor_3d[:, :, 2]  # 세 번째 채널 (10x10)

# 결과 출력
print("3D Tensor (10x10x3):")
print(tensor_3d)
print("\n2D Tensor 1 (10x10):")
print(tensor_2d_1)
print("\n2D Tensor 2 (10x10):")
print(tensor_2d_2)
print("\n2D Tensor 3 (10x10):")
print(tensor_2d_3)
```

[4] 3의 배수로 구성된 원소갯수가 12개인 텐서를 생성하세요.

[7] 3의 배수로 구성된 원소갯수가 12개인 텐서를 생성하세요.
- 1D 텐서 생성 T1
- T1의 축 추가로 2D 텐서 생성 T2
- T1의 축 추가로 3D 텐서 생성 T3
- T1의 축 추가로 4D 텐서 생성 T4

- T4의 축 축소로 3D 텐서 생성 T5
- T3의 축 축소로 2D 텐서 생성 T6
- T2의 축 축소로 1D 텐서 생성 T7

```python


# 3의 배수로 구성된 12개의 원소를 가진 1D 텐서 생성
T1 = torch.arange(3, 37, 3)  # [3, 6, 9, ..., 36]
print("T1 (1D Tensor):")
print(T1)

# T1의 축 추가로 2D 텐서 생성 (3x4 형태)
T2 = T1.reshape(3, 4)  # 텐서 T1을 (3, 4) 형태로 변환
print("\nT2 (2D Tensor):")
print(T2)

# T1의 축 추가로 3D 텐서 생성 (2x2x3 형태)
T3 = T1.reshape(2, 2, 3)  # 텐서 T1을 (2, 2, 3) 형태로 변환
print("\nT3 (3D Tensor):")
print(T3)

# T1의 축 추가로 4D 텐서 생성 (2x2x2x3 형태)
T4 = T1.reshape(1, 2, 2, 3)  # 텐서 T1을 (2, 2, 2, 3) 형태로 변환
print("\nT4 (4D Tensor):")
print(T4)

# T4의 축 축소로 3D 텐서 생성
T5 = T4.squeeze(3)  # 3번째 축을 제거하여 3D 텐서로 변환
print("\nT5 (3D Tensor):")
print(T5)

# T3의 축 축소로 2D 텐서 생성
T6 = T3.squeeze(2)  # 2번째 축을 제거하여 2D 텐서로 변환
print("\nT6 (2D Tensor):")
print(T6)

# T2의 축 축소로 1D 텐서 생성
T7 = T2.squeeze(1)  # 1번째 축을 제거하여 1D 텐서로 변환
print("\nT7 (1D Tensor):")
print(T7)
```

[문제 4] 2개의 텐서를 생성하고, 두 텐서의 덧셈을 수행해 보세요. 텐서 크기는 각각 (3, 3)입니다.

```python
t1 = torch.Tensor([[2, 2, 1], [3, 3, 1], [4, 4, 1]])
torch.manual_seed(0)
t2 = torch.arange(2, 11, 1)
t2 = t2.reshape(3, 3)

# torch.arange로 더 많은 값을 생성하여 크기 맞추기: 만약 (3, 3) 형태로 변형하려면, torch.arange로 9개의 요소를 생성해야 합니다

# 덧셈 연산
result = t1 + t2

print(t1.size())  # 텐서 t1의 크기 출력
print(t2.size())         # 텐서 t2 출력
print(result)     # 덧셈 결과 출력
```
