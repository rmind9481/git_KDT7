[1] 모듈로딩 및 데이터 준비 <hr>

```python
# [1-1] 모듈로딩
import pandas as pd
```

```python
# [1-2] 데이터 준비
DATA_FILE = '../DATA/member.csv'
```

```python
# [2-1] 데이터 로딩
memberDF = pd.read_csv(DATA_FILE)
```

```python
# [2-2] 데이터 확인
# -요약 정보 확인 => DF객체변수명.info()

memberDF.info() 
print(f"index {memberDF.index}")
print(f"columns {memberDF.columns}")
print(f"values {memberDF.values}")

print(f"head(){memberDF.head()}")
print(f"tail(){memberDF.tail()}")
print(f"type {type(memberDF)}")
print(f"shape {memberDF.shape}")
```

```python
# 실제 데이터 확인 => DF객체변수명.head(), DF객체변수명.tail() 

memberDF.head(2) 
```

```python
# 컬럼별 통계치 확인 => DF객체변수명.describe()
memberDF.describe(include='all')
```

[3] 행추가 <hr>

```python
# DF객체변수명.loc[추가 행 인덱스] = 값

memberDF.loc[7]=100
```

```python
memberDF
```

```python
memberDF.tail()
```

```python
# DF객체변수명.loc[추가 행 인덱스] = [컬럼수만큼]
memberDF.loc[8]=[9,'태권브이',100,'대구']
```

```python
memberDF.tail(2)
```

```python
# DF객체변수명.loc[추가 행 인덱스] = [컬럼수만큼]
# -> 컬럼수와 데이터가 다르면 ERROR
# memberDF.loc[9]=[9,'태권브이']
```

```python
# DF객체변수명.loc[추가 행 인덱스] = [컬럼수만큼]
# -> 동일 행인덱스 지정 시에 데이터가 변경됨
memberDF.loc[8]=0
memberDF.tail(2)
```

[4] 열추가 <Hr>

```python
# => DF객체변수명[컬러명] = 값
memberDF['성별'] = '남'
memberDF.head()
```

```python
# => DF객체변수명[컬러명] = [행갯수만큼]
memberDF['혈액형'] = ['A',"B","AB","O","B","A","AB","A","B"]
memberDF.head()
```

[DataFrame에서 원소 값 변경]<Hr> 

```python
# 홍길동 지역 변경 대구 => 부산
# -행 0번행 -- 열 지역

memberDF.loc[0, '지역'] = '부산'
memberDF.head()
```

```python
# 마징가, 박지민 지역 부산/대구 캐나다
# -행 0번행 -- 열 지역

memberDF.loc[[3,4], '지역'] = '캐나다'

memberDF.head()

memberDF.loc[[3,4], '지역'] = ['미국','프랑스']

memberDF.head()
```
