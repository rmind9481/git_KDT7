```python
## [1-1] 모듈 로딩
import pandas as pd                 # 데이터 분석 및 전처리 모듈
import matplotlib.pyplot as plt     # 시각화 모듈
```

```python
# [1-2] 데이터 준비

data = {'A':[10,20,30,40,50],
        'B':[15,25,35,45,55]}

dataDF = pd.DataFrame(data)
```

[2] 데이터시각화 -선 그래프
<hr>

```python
# => plt.plot(ydata) , plt.plot(xdata, ydata)
#  - A 컬럼 시각화 

yData = dataDF['A']
plt.plot(yData)
plt.show()
```

```python
# => plt.plot(ydata) , plt.plot(xdata, ydata)
#  - A 컬럼 시각화 

yData = dataDF['A']
xData = dataDF.index +1


plt.plot(xData, yData)
plt.show()
```

```python
# - 선X, 마크만 표시
# plt.plot(xData, yData, 'o' )     # 기본색상 o마커  
# plt.plot(xData, yData, 'ro' )    # 기본색상 ro마커 r 레드, 
plt.plot(xData, yData, 'm^' )    # magenda 색상 < 마커 
plt.show()
```

```python
#  - A 컬럼 시각화 

yData = dataDF['A']
xData = dataDF.index +1

# - 다양한 선 스타일, 마크 표시
# plt.plot(xData, yData, 'm*--') # magenda색상 * 마커 -- 스타일 선
#plt.plot(xData, yData, 'm*:') # magenda색상 * 마커 : 스타일 선
plt.plot(xData, yData, 'm*-.') # magenda색상 * 마커 -- 스타일 선

plt.show()
```

```python
#  - A 컬럼 시각화 

yData = dataDF['A']
xData = dataDF.index +1

# - 다양한 선 스타일, 마크, 선 두깨
# - lw 또는 lineWidth = 실수값
# plt.plot(xData, yData, 'm*--', lw=8.5)              # 선두깨 lw 설정

# - 마커 두께 설정 makerSize 또는 ms = 실수값
plt.plot(xData, yData, 'r*--', ms= 10)              # 마커 두깨

plt.show()
```

```python
#  - A 컬럼 B컬럼 시각화 

xData = dataDF.index +1

yData1 = dataDF['A']
yData2 = dataDF['B']

# -시각화 
plt.plot(xData,yData1, 'r*--', lw =2.0, ms=10)
plt.plot(xData,yData2)
plt.show()
```

```python
#  - A 컬럼 B컬럼 시각화  + 범례 

xData = dataDF.index +1

yData1 = dataDF['A']
yData2 = dataDF['B']

# -시각화 
plt.plot(xData,yData1, label = 'A')     # A 칼럼
plt.plot(xData,yData2, label = 'B')     # B 칼럼
# plt.legend()                            # 그래프별 라벨 설정 즉, 범례
#plt.legend(loc= 'upper center')
plt.legend(loc= 'center left')
plt.show()
```

```python
#  - 그래프에 제목& x 축이름 , y축이름

xData = dataDF.index +1

yData1 = dataDF['A']
yData2 = dataDF['B']

# -시각화 
plt.plot(xData,yData1, label = 'A')     # A 칼럼
plt.plot(xData,yData2, label = 'B')     # B 칼럼
plt.legend()                            # 그래프별 라벨 설정 즉, 범례
plt.title('Acolum & Bcolunm', pad=50,
        fontdict={'color':'r', 'size':'x-large', 'weight':'bold'} )         # 그래프 제목
plt.xlabel('Index', labelpad=15,  fontdict={'color':'b'})                    # X축 이름
plt.ylabel('Value',  labelpad=25)                        # Y축 이름
plt.grid()
plt.show()
```
