[3] 데이터 추출 <hr>

```python
# - 행 추출

dataDF.loc['제주특별자치도'] 
```

```python
## - '제주특별자치도' ==> '대구광역시' 

dataDF.loc[('제주특별자치도','대구광역시')]
 
```

```python
# - 전입지별이 '대구광역시' ===> 멀티인덱스의 Level 별 데이터 추출 => xs()메서드

dataDF.xs('대구광역시') # == dataDF.loc['대구광역시']
dataDF.xs('대구광역시', level='전입지별')


```

```python
d = {'num_legs': [4, 4, 2, 2,4],
      'num_wings': [0, 0, 2, 2,2],
      'class': ['mammal', 'mammal', 'mammal', 'bird','bird'],
      'animal': ['cat', 'dog', 'bat', 'penguin','cat'],
      'locomotion': ['walks', 'walks', 'flies', 'walks','flies']}


df = pd.DataFrame(data=d)

df = df.set_index(['class', 'animal', 'locomotion'])

df
```

```python
df.loc[('mammal','cat','walks')]
```

```python
df.xs('cat', level=1)
```

```python
df.xs('walks', level=2)
```

```python
 # - num_tails 컬럼추가
df['num_tails'] = 0
```
