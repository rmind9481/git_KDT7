```python
# [1-2] 데이터 불러오기

data = '../DATA/시도별 전출입 인구수.xlsx'

dataDF = pd.read_excel(data)
```

[3] 분석을 위한 데이터 가공 <hr>

```python
# 제거 : 0번행 제거
#  전출지별 => 진행의 값 결측치 채우기

#dataDF['전출지별'].fillna(method='ffill')
dataDF['전출지별'].ffill(inplace=True)

dataDF.head()
# FutureWarning => 미래에 일어날수있는 에러
```

```python
## '대구광역시' 행만 추출

# 대구광역시만 추출하는 확인용

#(dataDF['전출지별'] == '대구광역시').sum()


deaguIndex = (dataDF['전출지별'] == '대구광역시')

deaguDF = dataDF[deaguIndex]
```

```python
deaguDF
```

```python
# 대구 데이터프에림에서 필요없는 전출지별 컬럼 삭제
# -컬럼 : 전출지별
# -행   : 전입지별 전국, 대구 광역시 => 55,58
# [경고] SettingWithCopyWarning => iinplace = True 한 경우 
deaguDF = deaguDF.drop(columns='전출지별', index=[55,58]).copy()
```

```python
deaguDF.head()
```

```python
# '전입지별' 컬럼을 행 인덱스 설정
deaguDF.set_index('전입지별', inplace=True)
```

```python
deaguDF
```

```python
# ===> 데이터 가공
deaguDF.replace('-', '0', inplace=True)
```

```python
# ===> 모든 컬럼의 타입을 Object => int 변환
deaguDF = deaguDF.astype(int)
```

```python
deaguDF.info()
# 타입은 컬럼 단위로 잡힌다.
```

```python
# 년도별 전출자수 변화 분석 ==> 선 그래프 시각화

xData = deaguDF.columns
yData = deaguDF.loc['서울특별시']

# 그래프 들어갈 도화지/캔버스
plt.figure(figsize=(10,5))
for city in deaguDF.index:
    print(city)
    yData = deaguDF.loc[city]
    plt.plot(xData,yData, label = city)

plt.legend(loc = 'center left', bbox_to_anchor=(1,0.5))
plt.xticks(np.arange(0,50,10))
plt.show()
```
