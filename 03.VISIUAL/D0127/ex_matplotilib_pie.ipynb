[matplotlib 의 pie 시각화]
- 전체 대비 항목에 대한 비중/비율 시각화 그래프
- 데이터 제공 시 자동 계산 수행

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 준비
labels = 'Frogs', 'Hogs','Dogs',"Logs"
size = [15,30,45,10]

# 시각화
plt.pie(size , labels= labels, labeldistance=0.5)
plt.show()
```

```python
# 조각별 비율 출력 => autopct 매개변수 : 00.00%

plt.pie(size, labels=labels, autopct='%1.2f%%')
plt.show()
```

```python
# 커스텀 색상 설정 => colors 매개변수 : [ 조각별 원하는 색상 리스트]
plt.pie(size, labels=labels,
        colors=['olivedrab','rosybrown','saddlebrown','gray'])
```

```python
# 라벨 과 퍼센트 출력 위치 설정
# => pctdistance 매개변수 : 퍼센트 값 출력 위치 설정 [기]0.6
# => labeldistance 매개변수 : 라벨 출력 위치 설정 [기]1.1

plt.pie(size, labels=labels, autopct='%1.1f%%',
        pctdistance=1.25, labeldistance=.6)
plt.show()
```

```python
# 전체에서 일부 조각을 분리하는 설정
# explode 매개변수: 조각 수 만큼 중심점에서의 간격
# shadow  매개변수 : 그림자 설정

explode = (0.25,0.05,0.25,1.0) # only "explode" the 2nd slice(i.e, 'Hongs')

plt.pie(size, explode= explode,
        labels=labels,
        autopct='%1.5f%%',
        shadow=True,
        startangle=90)
```

```python
# 전체에서 일부 조각을 분리하는 설정
# explode 매개변수: 조각 수 만큼 중심점에서의 간격
# shadow  매개변수 : 그림자 설정
# 조각을 배치 방향 설정 counterclock=False 매개변수 : 기본값 반시계 방향 True
# 조각 배치 시작 각도 설정 startangle=90 방향설정  매개변수 : 기본값 0
explode = (0.25,0.05,0.25,1.0) # only "explode" the 2nd slice(i.e, 'Hongs')

plt.pie(size,labels=labels, counterclock=False, startangle=90)
plt.show()
```
