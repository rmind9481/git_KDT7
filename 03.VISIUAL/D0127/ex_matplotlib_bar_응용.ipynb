[ Matplotlib 데이터 시각화 - 막대 그래프 ]
- 항목/그룹간 비교에 사용되는 그래프

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## [1-1] 모듈 로딩
import pandas as pd                     # 데이터 분식 및 전처리 모듈
import matplotlib.pyplot as plt         # 시각화 모듈
```

```python
## [1-2] 데이터 준비 
data = {'A' :[10,20,30,40,50], 'B':[15,25,35,45,55]}

dataDF=pd.DataFrame(data)
```

```python
dataDF
```

[2] 데이터 시각화 - 막대 그래프 <hr>

```python
## - A컬럼 시각화 
xData = dataDF.index + 1
yData = dataDF['A']

## - 막대 그래프 시각화
plt.bar(xData, yData)

plt.title("A COLUMN")
plt.xlabel("INDEX")
plt.ylabel("VALUE")
plt.legend(labels=['A'])

## - 화면 출력
plt.show()
```

```python
## - A컬럼 시각화 
xData = dataDF.index + 1
yData = dataDF['A']

## - 막대 그래프 시각화
plt.bar(xData, yData, color='orange', align='edge')

plt.title("A COLUMN")
plt.xlabel("INDEX")
plt.ylabel("VALUE")
plt.legend(labels=['A'])

## - 화면 출력
plt.show()
```

```python
dataDF.sum()
```

```python
## - A컬럼과 B컬럼 시각화
sumSR = dataDF.sum()
xData = sumSR.index
yData = [sumSR['A'], sumSR['B']]
colors= ['orange','blue']

## - 막대 그래프 시각화
plt.bar(xData, yData, color=colors, align='center')

plt.title("A & B")
plt.xlabel("ITEM")
plt.ylabel("TOTAL")

plt.show()
```

```python
## - A컬럼과 B컬럼 시각화
sumSR = dataDF.sum()
xData = sumSR.index
yData = [sumSR['A'], sumSR['B']]
colors= ['orange','blue']

## - 막대 그래프 시각화
plt.bar(xData, yData, color=colors, width=0.4)

plt.xticks(xData, ['AAA','BBB'])

plt.title("A & B")
plt.xlabel("ITEM")
plt.ylabel("TOTAL")

plt.show()
```

```python
## - A컬럼과 B컬럼 시각화
sumSR = dataDF.sum()
xData = sumSR.index
yData = [sumSR['A'], sumSR['B']]
colors= ['orange','blue']

## - 막대 그래프 시각화
bar=plt.bar(xData, yData, color=colors, width=1.)

# 숫자 넣는 부분
# ha: 수평 정렬 설정 ('center', 'left', 'right')
# va: 수직 정렬 설정 ('center', 'top', 'bottom')

for rect in bar:
    h = rect.get_height()
    w = rect.get_width()
    x = rect.get_x()
    print(f'h => {h}, w => {w}, x => {x}')
    plt.text(x + w/2.0, h, f'{h:.1f}', ha='center', va='center', size = 12)


plt.xticks(xData, ['AAA','BBB'])

plt.title("A & B")
plt.xlabel("ITEM")
plt.ylabel("TOTAL")

plt.show()
```

```python
## - A컬럼과 B컬럼 시각화
sumSR = dataDF.sum()
xData = sumSR.index
yData = [sumSR['A'], sumSR['B']]
colors= ['green','pink']

## - 막대 그래프 시각화
barh=plt.barh(xData, yData, color=colors, height=0.4)

plt.yticks(xData, ['AAA','BBB'])

# ha: 수평 정렬 설정 ('center', 'left', 'right')
# va: 수직 정렬 설정 ('center', 'top', 'bottom')
for rect in barh:
    h = rect.get_height()
    w = rect.get_width()
    x = rect.get_x()
    y = rect.get_y()
    print(f'h => {h}, w => {w}, x => {x}, y = {y}')
    plt.text(x + w/2.0, y + h/2.0, f'{h:.1f}', ha='center', va='center', size = 12)

plt.title("A & B")
plt.xlabel("ITEM")
plt.ylabel("TOTAL")

plt.show()
```

막대그래프 응용 <hr>

```python
# 데이터 준비
w = 0.8
xData = pd.Series([10,10+w])
yData = pd.Series([40,60])

plt.bar(xData,yData, color=['red','green'])
plt.show()
```

```python
# 데이터 준비
xData = pd.Series(range(1,7,2))
yData1 = pd.Series([50,40,90])
yData2 = pd.Series([80,90,60])
yData3 = pd.Series([20,50,30])

#시각화 준비
w, w2 = 0.4, 0.4*2

plt.bar(xData, yData1, width=0.4)
plt.bar(xData+w, yData2, width=0.4)
plt.bar(xData+w2, yData3, width=0.4)

plt.xticks(xData+w, ['1970','1980','1990'])
plt.show()
```
