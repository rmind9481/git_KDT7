```python
# [2-2] auto_mpg.csv 파일 기반 산점도
#  - mpg 와 weight 컬럼의 관계성

data = '../DATA/auto_mpg.csv'

dataDF = pd.read_csv(data)
dataDF
```

```python
dataDF.info()
```

```python
dataDF.head()
```

```python

xData = dataDF['mpg']
yData = dataDF['weight']

plt.scatter(xData,yData, color ="red")
plt.xlabel("mpg")
plt.ylabel("weight")
plt.show()
```

```python
# -상관계수 값 추출
corrSR = dataDF.corr(numeric_only=True)['mpg']
corrSR
```

```python
# 관계성 시각화
plt.scatter(xData,yData, color ="hotpink")

plt.title(f" MPG& WEIGHT CORR VALE {corrSR['weight']:.2f}")
plt.xlabel("mpg")
plt.ylabel("weight")

plt.show()
```

```python
#  - mpg 와 weight , clydiners 컬럼의 관계성

print(dataDF.columns)


xData = dataDF['mpg']
yData = dataDF['weight']
sData = dataDF['cylinders']*12.5
# colors = ['hotpink','red','green']


# plt.scatter(xData,yData,sData, color='hotpink')
plt.scatter(xData, yData, sData, c=dataDF['cylinders'], cmap='viridis', alpha=0.7)
plt.title(f" MPG& WEIGHT & {corrSR['weight']:.2f} / MPG & CYLINDERS {corrSR['cylinders']:.2f}")
plt.xlabel("mpg")
plt.ylabel("weight")

plt.show()





```

```python
# mpg 와 origin 관계

xData = dataDF['mpg']
yData = dataDF['origin']

# 오리진의 값을 컬러값으로 변환후 저장
cData = dataDF['origin'].replace({1:'#86E57F',2:'#FFB2F5',3:'#F2CB61'})

print(corrSR['origin'])
plt.scatter(xData, yData , c=cData, alpha=0.7)

# plt.scatter(xData, yData , c=dataDF['origin'], cmap='cool', alpha=0.7)
plt.title(f" MPG & Origin {corrSR['origin']:.2f}")
plt.xlabel("mpg")
plt.ylabel("origin")

plt.show()

```
