[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# [1-1] 모듈 로딩

import pandas as pd
import matplotlib.pyplot as plt
import utils
```

```python
# [1-2] 데이터 준비

xData = pd.Series([1,2,3,4])
yData = pd.Series([2,3,5,10])

# 시각화
plt.plot(xData,yData)
plt.xlabel("X-Axis")
plt.xlabel("Y-Axis")


# 영역 채우기
plt.fill_between(xData[1:3],yData[1:3], alpha = 0.5)
# 그래프 그리고
plt.show()
```

```python
xData = pd.Series([1,2,3,4])
yData = pd.Series([2,3,5,10])


plt.plot(xData,yData)

plt.fill_betweenx(yData[2:4],xData[2:4], alpha=0.3)
plt.grid()
plt.xlabel("X축")
plt.ylabel("y축")
plt.show()
```

```python
xData = pd.Series([1,2,3,4])
yData1 = pd.Series([2,3,5,10])
yData2 = pd.Series([1,2,4,8])


plt.plot(xData,yData1)
plt.plot(xData,yData2)

# 영역 채우기 - 2개 선 사이 채우기
plt.fill_between(xData[1:3],yData1[1:3],yData2[1:3], alpha=0.3)
plt.grid()
plt.xlabel("X축")
plt.ylabel("y축")
plt.show()
```

```python
xData = pd.Series([1,2,3,4])
yData1 = pd.Series([2,3,5,10])
yData2 = pd.Series([1,2,4,8])


plt.plot(xData,yData1)
plt.plot(xData,yData2)

# 영역 채우기 - 2개 선 사이 채우기
plt.fill_between(xData[2:4],yData1[2:4],yData2[2:4], alpha=0.3)
plt.grid()
plt.xlabel("X축")
plt.ylabel("y축")
plt.show()
```

```python
xData = pd.Series([1,2,3,4])
yData1 = pd.Series([2,3,5,10])
yData2 = pd.Series([1,2,4,8])


plt.plot(xData,yData1)
plt.plot(xData,yData2)

# 영역 채우기 - 다각형
plt.fill([1.7, 1.7 , 2.25 ,3.0 ,3.0],[2.2, 3.5, 5.5, 4.8, 3.2], alpha=0.25)
plt.grid()
plt.xlabel("X축")
plt.ylabel("y축")
plt.show()
```
