[ Matplotlib의 히스토그램(histogram)]

- 데이터의 분포 체크위한 시각화 그래프
- 이상치/패턴 파악 가능함

[1] 모듈로딩 <hr>

```python
import matplotlib.pyplot as plt
import numpy as np
```

[2] 데이타 준비 <hr>

```python
# 난수 추출 고정 
np.random.seed(3)

# 정규분포 형태의 임의의 데이터 1000개 추출
data = np.random.randn(1000)

print(f'평균: {data.mean():.1f}, 표준편차 : {data.std():.1f}')
```

```python
# 시각화
# 구간 설정

plt.hist(data, label='Biins 10')
plt.legend()
plt.show()
```

```python
# 시각화
# bins 매개변수 : 막대수 설정

plt.hist(data, bins=30, label='Bins 30', edgecolor= 'black', linewidth = 0.5)
plt.legend()
plt.show()
```

```python
# 시각화
# bins 매개변수 : 막대수 설정

fig = plt.figure(figsize=(3,3))
fig.set_facecolor('lightgray')
plt.hist(data, bins=30, label='Bins 30',
         color='green', edgecolor='black')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(3,3))
fig.set_facecolor('lightgray')
plt.hist(data, bins=30, label='Bins 30',
         color='green', edgecolor='black',
         histtype='step')
plt.legend()
plt.show()
```

```python
# 난수 추출 고정 
np.random.seed(3)

# 정규분포 형태의 임의의 데이터 1000개 추출
data2 = (np.random.randn(1000)+2) *1.5

print(f'평균: {data2.mean():.1f}, 표준편차 : {data2.std():.1f}')

fig = plt.figure(figsize=(5,5))
fig.set_facecolor('lightgray')
# density 매개변수 : y축을 비율로 표현
plt.hist(data,  label='Data1', density=True,
        alpha=0.75,  edgecolor='k')
plt.hist(data2,  label='Data2', density=True,
        alpha=0.75,  edgecolor='k')

plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
fig.set_facecolor('lightgray')


# density 매개변수 : y축을 비율로 표현
plt.hist([data,data2],color=['green','blue'] ,
         histtype='barstacked',
         label=['Data1','Data2'],
         density=True, alpha=0.75,edgecolor='k')
plt.legend()
plt.show()
```
