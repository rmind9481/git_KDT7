[ Matplotlib의 여러개 그래프를 하나의 Figure에 그리기]

- subplot()
- subplots()

[1] 모듈로딩 <hr>

```python
import matplotlib.pyplot as plt
import pandas as pd
```

[2] 데이타 준비 <hr>

```python
## 파일 준비

D_FILE = '../DATA/auto_mpg.csv'

autoDF = pd.read_csv(D_FILE)
autoDF.head()
```

[3] 시각화 <hr>

```python
# 여러개 그래프 그리기 - 방법 1 ) subpolt(행, 열, 인덱스)


fig = plt.figure()
fig.set_facecolor('hotpink')

# 전체제목
plt.suptitle('[MPY & 4-GRAPHS]')

plot1 = plt.subplot(2,2,1)
plot1.plot(autoDF['mpg'], 'b-')
plot1.set_title('MPG&LINE')

plot2 = plt.subplot(2,2,2)
plot2.bar(autoDF.index, autoDF['mpg'])
plot2.set_title("MPG&BAR")

plot3 = plt.subplot(2,2,3)
plot3.hist(autoDF['mpg'], histtype='step')
plot3.set_title('MPG & HISTOGRAM')

plot4 = plt.subplot(2,2,4)
plot4.boxplot(autoDF['mpg'], notch=True, patch_artist=True,
              boxprops=dict(color='yellow', facecolor='red'))
plot4.set_title('MPG & BOXPLOT')

plt.tight_layout()
plt.show()
```

```python
#
nrow, ncols = 2,5

fig = plt.figure(figsize=(16,8))
fig.set_facecolor('hotpink')

# 전체제목
plt.suptitle('[MPY & 4-GRAPHS]')

plot1 = plt.subplot(nrow, ncols,1)
plot1.plot(autoDF['mpg'], 'b-')
plot1.set_title('MPG&LINE')

plot2 = plt.subplot(nrow, ncols,2)
plot2.bar(autoDF.index, autoDF['mpg'])
plot2.set_title("MPG&BAR")

plot3 = plt.subplot(nrow, ncols,3)
plot3.hist(autoDF['mpg'], histtype='step')
plot3.set_title('MPG & HISTOGRAM')

plot4 = plt.subplot(nrow, ncols,4)
plot4.boxplot(autoDF['mpg'], notch=True, patch_artist=True,
              boxprops=dict(color='yellow', facecolor='red'))
plot4.set_title('MPG & BOXPLOT')

plt.tight_layout()
plt.show()
```

```python
# 여러개 그래프 그리기 -방법 2) subplots(행,열) => figure 객체와 axes 객체 리스트
nrow, ncols = 2,4

fig, axes = plt.subplots(nrow,ncols)


# print(f'axes => {axes}')
# axes[0,0].plot(autoDF['mpg'], 'ro')
# axes[1,2].plot(autoDF['mpg'], 'yo')

for row in range(nrow):
    for col in range(ncols):
        axes[row,col].plot(autoDF['mpg'], 'ro')



plt.suptitle("All TITLE : 6-PLOTS")
plt.tight_layout()
plt.show()
```

```python
# 여러개 그래프 그리기 -방법 2) subplots(행,열) => figure 객체와 axes 객체 리스트
nrow, ncols = 2,2
colnames = ['cylinders','displacement','weight','origin']
fig, axes = plt.subplots(nrow,ncols, sharex=True)


# print(f'axes => {axes}')
# axes[0,0].plot(autoDF['mpg'], 'ro')
# axes[1,2].plot(autoDF['mpg'], 'yo')

for row in range(nrow):
    for col in range(ncols):
        axes[row,col].scatter(autoDF['mpg'], autoDF[colnames[col+row*2]])
        axes[row,col].set_title(f"['mpg'& {colnames[col+row*2]}]")



plt.suptitle("All TITLE : 6-PLOTS")
plt.tight_layout()
plt.show()
```

```python
# 여러개 그래프 그리기 -방법 2) subplots(행,열) => figure 객체와 axes 객체 리스트
nrow, ncols = 2,2
colnames = ['cylinders','displacement','weight','origin']
fig, axes = plt.subplots(nrow,ncols, sharex=True)

# 2D => 1D 리스트
print(axes.flatten())

for ax, col_name in zip(axes.flatten(), colnames):
    ax.scatter(autoDF['mpg'], autoDF[col_name])
    ax.set_title(f"['mpg'& {col_name}]")

plt.suptitle("All TITLE : 6-PLOTS")
plt.tight_layout()
plt.show()
```
