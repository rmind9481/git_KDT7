```python
# 모듈및 데이터 준비
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
data = pd.read_excel("../DATA/남북한발전전력량.xlsx")

power_DF = pd.DataFrame(data)
```

```python
power_DF
```

```python
power_DF.info()
```

```python
# 전력량 남한 북한 추출

power_DF['전력량 (억㎾h)'].ffill(inplace=True)
```

```python
# 데이터확인

power_DF.head(5)
```

```python
df1 = power_DF[power_DF['전력량 (억㎾h)'] == '남한']
df2 = power_DF[power_DF['전력량 (억㎾h)'] == '북한']
```

```python
df1.head()
```

```python
df2.head()
```

```python
# 남한, 북한 전력량 컬럼 삭제, 합계 삭제

df1.columns
df1.drop(columns=["전력량 (억㎾h)"], index=[0,4], inplace=True)

# df1_Noth = df1.drop([0]).copy()

# print(type(df1_Noth))
df1.set_index("발전 전력별", inplace=True)

df1
```

```python
# 북한 전처리
# 수력, 화력만 추출
df2.drop(columns=["전력량 (억㎾h)"], index=[5,8], inplace=True)
df2.set_index("발전 전력별", inplace=True)
df2
```

```python
# object => int 형변환

# df1 = df1.astype(int)
# df2 = df2.astype(int)
df1.values.astype(int)
df2.values.astype(int)
```

```python
df1.info()
df2.info()
```

```python
## 그래프 
# X 축은 년도
# Y 축은 에너지 
year = []

for i in df1.columns:
    year.append(int(i))

print(year)


xData = year

```

```python

for i in df1.index :
    yData = df1.loc[i]
    plt.plot(xData, yData, label=f'남한 {i}')
for i in df2.index :
    yData = df2.loc[i]
    plt.plot(xData, yData, label=f'북한 {i}')
plt.legend()
plt.show()
```

# 막대바 그래프 

```python
# X 축년도  y 축 에너지 명 


for i in df1.index :
    yData = df1.loc[i]
    plt.bar(xData, yData, label=f'남한 {i}')
for i in df2.index :
    yData = df2.loc[i]
    plt.bar(xData, yData, label=f'북한 {i}')

plt.legend()
plt.show()
```
