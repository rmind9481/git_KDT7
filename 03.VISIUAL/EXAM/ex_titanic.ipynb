```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

data = "../DATA/titanic.csv"

titanic_DF = pd.read_csv(data)
```

```python
titanic_DF.info()
```

```python
titanic_DF.head()
```

```python
plt.figure(figsize=(10,5))

titanic_DF['survived'].value_counts()
plt.subplot(1,2,1)

titanic_DF['survived'].value_counts().plot(kind='bar')
plt.subplot(1,2,2)

# pie 그래프
titanic_DF['survived'].value_counts().plot(kind='pie', autopct = '%1.1f%%')

```

```python
plt.subplot(1,2,1)

# 성별 분포에 대한 막대그래프

titanic_DF['sex'].value_counts().plot(kind = 'bar')

plt.subplot(1,2,2)

sns.countplot( data=titanic_DF, x = titanic_DF['sex'], hue= titanic_DF['survived'])
```

```python
titanic_DF['sibsp'].value_counts().plot(kind = "bar")
```

```python
plt.subplot(1,2,1)
titanic_DF['embarked'].value_counts().plot(kind = "bar")
plt.subplot(1,2,2)
sns.countplot(data=titanic_DF, x='embarked', hue="survived")
```

```python
titanic_DF['age'].plot(kind='hist')
```

```python
age_before = titanic_DF[titanic_DF['age'].isnull()]
age_before.head()
```

```python
mean_age = titanic_DF['age'].mean()
mean_age
```

```python
titanic_DF['age'] = titanic_DF['age'].fillna(mean_age)
titanic_DF[titanic_DF.isin(age_before.index)]

titanic_DF[titanic_DF['age'].isnull()]

titanic_DF['age'].plot(kind='hist')
```

```python
titanic_DF
```

```python
titanic_DF['embarked'].value_counts().plot(kind='bar')

em_before = titanic_DF[titanic_DF['embarked'].isnull()]
em_before.head()
```

```python
titanic_DF[['embarked']].mode()
```

```python
# 최빈값('S')로 Embarked의 null값 대체
mode_em = titanic_DF['embarked'].mode
titanic_DF['embarked'] = titanic_DF['embarked'].fillna('S')

# 결측치 채운 후
# Embarked의 값이 null이었던 data 조회
titanic_DF[titanic_DF.index.isin(em_before.index)]
```

```python
# 나이 분포도
plt.figure(figsize=(8,5))
sns.displot(titanic_DF['age'], bins =25)
```

```python
# 나이가 생존 여부에 영향을 끼치는지 확인인
sur = titanic_DF[titanic_DF['survived'] == 1]
dead = titanic_DF[titanic_DF['survived'] == 0]

plt.figure(figsize=(8, 5))
sns.kdeplot(x=sur['age'])
sns.kdeplot(x=dead['age'])
```
