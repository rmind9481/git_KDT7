```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.parse import quote

query	= quote(f'대구+날씨')
url = 'https://search.naver.com/search.naver?query=' +	query

html = urlopen(url)
soup = BeautifulSoup(html.read(), 'html.parser')

```

```python
print(soup)
```

```python
# 중구 동인동1가 기준 찾기
daegu_area = soup.find('span',{'class':'select_txt_sub'})
```

```python
daegu_area_list = daegu_area.text[:-2]
```

```python
print(f"대구광역시: {daegu_area_list}")
```
