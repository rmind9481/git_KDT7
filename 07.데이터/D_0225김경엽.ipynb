```python
# 패키지들을 임포트합니다.
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

# 영어와 수학 점수를 배열로 만듭니다.
english_scores = np.array([80,50,90,55,100,85,98,76,66,78,64,78])
math_scores = np.array([45,23,44,38,95,88,85,89,100,97,100,100])

# 영어 점수를 x값으로, 수학 점수를 y값으로 하여 1차 다항식을 이용해 데이터에 맞는 직선을 찾습니다.
poly_fit = np.polyfit(english_scores, math_scores, 1)

# 찾은 다항식 계수로 다항식 객체를 만듭니다.
poly_1d = np.poly1d(poly_fit)

# x값의 범위를 정합니다.
xs = np.linspace(english_scores.min(), english_scores.max())

# x값에 대한 y값을 계산합니다.
ys = poly_1d(xs)

# 그래프를 그리기 위한 figure와 subplot을 만듭니다.
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)

# 계산한 직선 그래프를 그립니다.
ax.plot(xs, ys, color='gray', label=f'{poly_fit[1]:.2f} + {poly_fit[0]:.2f}x')

# 실제 영어 점수와 수학 점수 데이터를 산점도로 그립니다.
ax.scatter(english_scores, math_scores, label='score')

# x축과 y축의 라벨을 설정합니다.
ax.set_xlabel('english')
ax.set_ylabel('mathematics')

# 범례를 그래프의 왼쪽 위에 표시합니다.
ax.legend(loc='upper left')

# 그래프를 화면에 표시합니다.
plt.show()
```

# 다차원 배열

### 배열 속성 확인

배열은 다양한 속성을 가지고 있으며, shape, ndim, dtype 등을 확인할 수 있습니다.


```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr.shape)  # (2, 3) -> (행, 열)
print(arr.ndim)   # 2 (차원 수)
print(arr.dtype)  # int64 (데이터 타입)
print(arr.size)   # 6 (원소 개수)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)  # [5, 7, 9] (벡터 덧셈)
print(a - b)  # [-3, -3, -3] (벡터 뺄셈)
print(a * b)  # [4, 10, 18] (원소별 곱)
print(a / b)  # [0.25, 0.4, 0.5] (원소별 나눗셈)
```

# 벡터 크기(Norm)

벡터의 크기는 각 원소의 제곱합의 제곱근입니다.

넘파이에서는 np.linalg.norm() 함수를 사용합니다.



단위 벡터(Unit Vector)

크기가 1인 벡터

벡터를 자기 자신의 크기로 나누면 단위 벡터가 됩니다.

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

dot_product = np.dot(v1, v2)  # 내적 계산
cross_product = np.cross(v1, v2)  # 외적 계산
```

# 두 벡터 간 거리

두 벡터의 차이를 계산한 후, np.linalg.norm()을 사용해 거리 측정

피타고라스 정리를 활용한 유클리드 거리 계산

```python
v1 = np.array([1, 2])
v2 = np.array([4, 6])

distance = np.linalg.norm(v1 - v2)  # 벡터 간 거리
distance
```

선형대수 (Linear Algebra)

✅ 벡터와 행렬

벡터(Vector) : 1차원 배열

행렬(Matrix) : 2차원 배열

텐서(Tensor) : 3차원 이상의 배열

```python
A = np.array([[1, 2], [3, 4]])  # 2x2 행렬
B = np.array([[5, 6], [7, 8]])  # 2x2 행렬
```
