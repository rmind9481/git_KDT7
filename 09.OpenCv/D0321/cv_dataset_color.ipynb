```python
# [1-1] 모듈 로딩

import cv2					# 일번적인 이미지 파일들에서 순수 이미지 데이터ㅐ 다루는 모듈
import matplotlib.pyplot as plt
import numpy as np			# 순수 이미지 데이터의 저장 형식 ndarray
import os
```

```python
# [1-2] 데이터 준비
IMG_DIR = '../image/cat_dog/' 	# 이미지 존재 폴더
ROW, COL = 50,50				# 이미지 리사이징 크기
DS_DIR = '../image/DS/'			# 학습에 사용할 이미지 저장 폴더
IMG_CSV = '../animal.csv'		# 이미지 라벨 과 이미지 로우 데이터 저장 파일
```

```python
# 학습용 이미지 저장 폴더 생성
if not os.path.exists(DS_DIR) :
	os.mkdir(DS_DIR)
else:
	print(f'{DS_DIR} 존재함')
```

```python
# [1-3] 사용 이미지 파일 리스트 추출

IMG_FILELIST = os.listdir(IMG_DIR)

DEBUG_MODE = True

for filename in IMG_FILELIST:
	if not os.path.isfile(IMG_DIR+filename):
		# 파일이 아닌 경우 처리
		IMG_FILELIST.remove(filename)
if DEBUG_MODE: print(IMG_FILELIST)
```

```python
for filename in IMG_FILELIST:
	# 파일명에서 확장자 추출
	#print(filename[ filename.index('.')+1: ])
	idxValue= filename.find('.')
	if idxValue > 0 :
		print(filename[ filename.index('.')+1: ])

	if filename[ idxValue+1: ] not in ['jpg','png','jpeg','bmp']:
				IMG_FILELIST.remove(filename)
	else:
		print(f'NOT FILE : {filename}')
		IMG_FILELIST.remove(filename)
		
print(IMG_FILELIST)
```

[2] 이미지 데이터 로딩 <hr>

```python
# Graysale 형식으로 이미지 데이터 로딩 

with open(IMG_CSV, mode='a') as csvF:

	for  filename in IMG_FILELIST:
		RGB_IMG = cv2.imread(IMG_DIR+filename, cv2.IMREAD_COLOR)
		print(f'RGB_IMG => {RGB_IMG.shape}')
	
		# - 사이즈 조정
		RGB_IMG = cv2.resize(RGB_IMG, (ROW, COL))
		print(f'RGB_IMG => {RGB_IMG.shape}')



		# - 파일로 저장
		resize_File = f'{DS_DIR}resize_{filename}'
		print(f'resize_File =>{resize_File}' )

		# open cv 체널 순서 BGR ==> 일반 RGB
		ret = cv2.imwrite(resize_File, cv2.cvtColor(RGB_IMG, cv2.COLOR_BGR2RGB))
		if not ret : print(f'fail => {resize_File}')

		# 체널 분리
		b,g,r = cv2.split(RGB_IMG)
		b,g,r = b.reshape(-1), g.reshape(-1), r.reshape(-1)

		# -csv 파일 쓰기 => 라벨, 픽셀1,.... 픽셀n
		label = filename[:3]
		data = np.concatenate((r,g,b))
		print(f'label ==> {label}, data => {data.shape}')

		sdata = list(map(lambda n : str(n), data))
		print(f'sdata => {sdata[:30]}')
		
		csvF.write(str(label)+',') # 숫자라벨쓰기
		csvF.write(','.join(sdata)+'\n') # 리스트 이미지 데이터 => 문자열 변환
		break



```
