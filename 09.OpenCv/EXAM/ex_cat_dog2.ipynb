```python
import cv2
import pandas as pd
import numpy as np
import struct 			
import os
```

```python
# 데이터 준비


IMG_DIR = '../image/cat_dog/'
file_list = [[],[]]

all_items = os.listdir(IMG_DIR)


for item in all_items:
	print(item)
	print(os.path.isfile(IMG_DIR+item))
	# 파일 과 폴더 체크후 파일만 추출
	
	if os.path.isfile(IMG_DIR+item) :
		
		if 'cat' in item.lower():
			cat = cv2.imread(IMG_DIR+item, cv2.IMREAD_GRAYSCALE)
			# 이미지 크기
			
		
			down_image = cv2.resize(cat,(50,50), interpolation=cv2.INTER_AREA)
			
			H , W = down_image.shape[:2]

			# 회전 중심 및 변환 
			M1 = cv2.getRotationMatrix2D((W//2,H//2),30,1.0)
			M2 = cv2.getRotationMatrix2D((W//2,H//2),45,1.0)
			M3 = cv2.getRotationMatrix2D((W//2,H//2),75,1.0)
			M4 = cv2.getRotationMatrix2D((W//2,H//2),90,1.0)
			M5 = cv2.getRotationMatrix2D((W//2,H//2),110,1.0)
			
			
			down_image0 = cv2.flip(down_image, 0)
			down_image1 =  cv2.flip(down_image, 1)
			down_image2 =  cv2.flip(down_image, -1)

			rotated_img1 = cv2.warpAffine(down_image  , M1, (W, H))
			rotated_img2 = cv2.warpAffine(down_image  , M2, (W, H))
			rotated_img3 = cv2.warpAffine(down_image  , M3, (W, H))
			rotated_img4 = cv2.warpAffine(down_image  , M4, (W, H))
			rotated_img5 = cv2.warpAffine(down_image  , M5, (W, H))

			file_list[0].extend([down_image,down_image0,down_image1,down_image2,
						rotated_img1,rotated_img2,rotated_img3,rotated_img4,rotated_img5])
		if 'dog' in item.lower():
			
			dog = cv2.imread(IMG_DIR+item,cv2.IMREAD_GRAYSCALE)
			
			
			down_image = cv2.resize(dog,(50,50), interpolation=cv2.INTER_AREA)

						
			H , W = down_image.shape[:2]

			# 회전 중심 및 변환 
			M1 = cv2.getRotationMatrix2D((W//2,H//2),30,1.0)
			M2 = cv2.getRotationMatrix2D((W//2,H//2),45,1.0)
			M3 = cv2.getRotationMatrix2D((W//2,H//2),75,1.0)
			M4 = cv2.getRotationMatrix2D((W//2,H//2),90,1.0)
			M5 = cv2.getRotationMatrix2D((W//2,H//2),110,1.0)
			


			down_image0 = cv2.flip(down_image, 0)
			down_image1 =  cv2.flip(down_image, 1)
			down_image2 =  cv2.flip(down_image, -1)
			
			rotated_img1 = cv2.warpAffine(down_image  , M1, (W, H))
			rotated_img2 = cv2.warpAffine(down_image  , M2, (W, H))
			rotated_img3 = cv2.warpAffine(down_image  , M3, (W, H))
			rotated_img4 = cv2.warpAffine(down_image  , M4, (W, H))
			rotated_img5 = cv2.warpAffine(down_image  , M5, (W, H))
			
			file_list[1].extend([down_image,down_image0,down_image1,down_image2
						,rotated_img1,rotated_img2,rotated_img3,rotated_img4,rotated_img5])

print(f' 이미지 파일 리스트 : {len(file_list)}개 \n{file_list}')
```

```python
# 개와 고양이 데이터 분리
cat_images = np.array(file_list[0])  # 고양이 이미지
dog_images = np.array(file_list[1])  # 개 이미지
```

```python
type(cat_images)
```

```python
# 라벨 생성 (고양이: 0, 개: 1)
cat_labels = np.zeros(len(cat_images), dtype=np.int8)
dog_labels = np.ones(len(dog_images), dtype=np.int8)
```

```python
from sklearn.utils import shuffle

# 데이터 합치기
X = np.vstack((cat_images, dog_images))  # 이미지 데이터 (고양이 + 개)
y = np.hstack((cat_labels, dog_labels))  # 정답 라벨 (0=고양이, 1=개)

# 데이터를 DataFrame으로 변환 후 섞기
df = pd.DataFrame(X.reshape(len(X), -1))  # 이미지 데이터를 1차원 벡터로 변환
df['label'] = y  # 라벨 추가

# 데이터 섞기
df = shuffle(df, random_state=42)

# 다시 NumPy 배열로 변환
X_shuffled = df.iloc[:, :-1].values  # 이미지 데이터
y_shuffled = df.iloc[:, -1].values   # 정답 레이블

print(f"랜덤 섞인 데이터 크기: {X_shuffled.shape}, 레이블 크기: {y_shuffled.shape}")
```

```python
from sklearn.model_selection import train_test_split

# 학습용 80%, 테스트용 20%로 데이터 나누기
X_train, X_test, y_train, y_test = train_test_split(X_shuffled, y_shuffled, test_size=0.2, random_state=42)

print(f"학습 데이터 크기: {X_train.shape}, 테스트 데이터 크기: {X_test.shape}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# KNN 모델 생성 및 학습
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# 예측 수행
y_pred_knn = knn.predict(X_test)

# 정확도 평가
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"KNN 모델 정확도: {accuracy_knn:.4f}")
```

```python
from sklearn.svm import SVC

# SVM 모델 생성 및 학습
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)

# 예측 수행
y_pred_svm = svm.predict(X_test)

# 정확도 평가
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM 모델 정확도: {accuracy_svm:.4f}")
```
