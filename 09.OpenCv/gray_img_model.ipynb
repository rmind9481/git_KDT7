```python
#[1-1] 모듈 로딩
import pandas as pd
import numpy as np

# 학습용/테스트용 데이터셋 분리
from sklearn.model_selection import train_test_split 
# 타겟/라벨 인코딩용
from sklearn.preprocessing import LabelEncoder
```

```python
# [1-2] 데이터 준비
IMG_DATA = './cat_dog.csv'
```

```python
# [1-3] 데이터 로딩
img_df = pd.read_csv(IMG_DATA, header=None)
img_df.info()
```

# [2] 이미지 데이터 전처리 <hr> 

```python
# [2-1] 피쳐 와 타겟 분리
img_df.head()
```

```python
# 타겟 추출
target_SR = img_df[0]
target_SR

# 피쳐 추출
# 2 차원
feature_DF = img_df[img_df.columns[1:]]

print(f' feature_DF : {feature_DF.shape}, tarfet_SR: {target_SR.shape}')
```

```python
# [2-2] 피쳐에 대한 전처리 => 이미지 상관관계 X, 데이터 타입, 값의 범위 맞추기
# 0~255  ==> 0.0 ~ 1.0 정규화
# => csv 파일에서 이미지 로딩 => st ==> unit8 
feature_DF = feature_DF.astype('uint8')
feature_DF.dtypes
```

```python
feature_DF.values
```

```python
# => 0 ~ 255 : 0.0 ~ 1.0 정규화
feature_DF = feature_DF/255.
feature_DF
```

```python
# [2-3] 타겟에 대한 전처리 => str/object 타입의 경우 수치화 인코딩

target_EN = LabelEncoder()
target_EN.fit(target_SR)
target_SR = target_EN.transform(target_SR)
```

```python
# -인코딩 된 정보
target_EN.classes_
```

```python
 # -타겟 데이터 확인
target_SR
```

```python
# 타겟 데이터 확인
print('타겟의 고유값', np.unique_values(target_SR))
```

# [학습 준비] <hr>

```python
# 학습용/테스트용 데이터셋 준비

X_train, X_test, y_train,y_test = train_test_split(feature_DF,
													target_SR,
													test_size=0.2,
													random_state=42,
													stratify=target_SR) # 분류의 경우 타겟 클래스 비율 유지 설정

```

```python
# 확인
print('[Train]', X_train.shape, y_train.shape)
print('[Test]', X_test.shape, y_test.shape)
```

# [4] 학습 진행 <hr>

```python
## [4-1] 분류 모델 관련 모듈로딩

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# [4-2] 모델별 학습 진행

lr_model = LogisticRegression(max_iter=10000)
lr_model.fit(X_train, y_train)
```

```python
# [5-1] 분류 모델 관련 검증 모델
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report

# 손실/비용 함수 : 손실계산  정답-예측 차이
from sklearn.metrics import log_loss
```

```python
# [5-2] 검증 진행 ==> 정답 - 예측
result_df = pd.DataFrame(columns=['model','kind','accuray','precision','recall','f1','loss'])

for kind, data, y_true in [('Train',X_train, y_train),('Test',X_test,y_test)]:
	pre_y = lr_model.predict(data)

	print(pre_y.shape, data.shape)
	print(pre_y)
	# 성능평가 : 정답 (1)  == 예측(1)  O, 정답 1 != 예측 (0) , X
	# 정확도
	accuracy = accuracy_score(y_true, pre_y)
	# 정밀도
	precision = precision_score(y_true, pre_y)
	# 재현율
	recall = recall_score(y_true, pre_y)
	# 확률 => 클래스
	f1 = f1_score(y_true, pre_y)

	# -손실/비용 계산 : 정답 1  ---> 1일 확률 0.23 : 1 - 0.23 = 0.77
	proba_y = lr_model.predict_proba(data)
	loss = log_loss(y_true,proba_y)

	# 결과 DataFrame 에 저장
	model_name = lr_model.__class__.__name__
	print(model_name)

	# 

	result_df.loc[result_df.shape[0]+1] = [model_name, kind, accuracy,precision,recall,f1,loss]

	# 전체 결과 출력
	print(f'accuracy : {accuracy:.4f} precision : {precision:.4f}, recall : {recall:.4f}, f1 : {f1:.4f}')
	print(f'loss : {loss:.4f}')
	
```
