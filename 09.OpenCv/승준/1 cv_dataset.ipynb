[1] 모듈 로딩 및 데이터 준비 

```python
import cv2					# 일반적인 이미지 파일들에서 순수 이미지 데이터 다루는 모델 
import matplotlib.pyplot as plt
import numpy as np			# 순수 이미지 데이터의 저장형식 > ndarrary
import os					# 폴더 파일 경로 
```

```python
# [1-2] 데이터 준비
IMG_DIR  = '../data/cat_dog/' # 이미지 가지고 올 곳 / 이미지 존재 폴더
ROW, COL = 60, 60 # < 이미지 용량 줄이기 위해 사용 / 이미지 리사이징 크기
IMG_CSV = '../data/animal.csv' # 이미지 저장 할 csv / 이미지 라벨과 이미지 로우데이터 저장하는 파일 
```

```python
# [1-3 사용 이미지 파일 리스트]
IMG_FILELIST = os.listdir(IMG_DIR) # 파일만 확인

# IMG_DIR + 'cat_(1).jpg' # 디버깅 통해서 파일명 제대로 불러왔는지 확인

for filename in IMG_FILELIST:
    #print(IMG_DIR +filename)
	if not os.path.isfile(IMG_DIR + filename):
		# 파일이 아닌 경우 리스트에서 제거
		IMG_FILELIST.remove(filename)

print(IMG_FILELIST)
```

```python
for filename in IMG_FILELIST:
    print(filename[filename.index('.')+1:]) # .의 위치 다음부터가 확장자 
    if filename[filename.index('.')+1:] not in ['jpg','jpeg','png','bmp']:
        IMG_FILELIST.remove(filename)
        
print(IMG_FILELIST)
```

[2] 이미지 데이터 로딩

```python
# grayscale 형식으로 이미지 데이터 로딩
with open (IMG_CSV, mode='a') as csvF:

	for filename in IMG_FILELIST:
		# 1 채널로 이미지 로딩
		grayIMG = cv2.imread(IMG_DIR+filename, cv2.IMREAD_GRAYSCALE)
		print(f"grayIMG > {grayIMG.shape}") # 디버깅용
		# 사이즈 조정
		grayIMG = cv2.resize(grayIMG, (ROW, COL))
		print(f"grayIMG.re > {grayIMG.shape}") # 디버깅용
		# - 파일로 저장
		grayFile = f"{IMG_DIR}gray_{filename}"
		print(f"grayFile > {grayFile}") # 디버깅용
		ret = cv2.imwrite(grayFile, grayIMG)
		if not ret: print(f"fali > {grayFile}")
		# csv파일에 쓰기
		
		label = filename[:3]
		data = grayIMG.reshape(-1) # csv에 저장하기 위해 차원 펴기
		print(f"label > {label}, data.shape > {data.shape}")

	
		sdata = list(map(lambda n: str(n), data))  # 문자열로 변환 (하나 하나를 문자열로 변환 후 list에 저장)
		print(f"sdata > {sdata[:30]}")
      
    	# CSV 파일에 쓰기 : 라벨, p1, p2, ...
		csvF.write(str(label)+",")                 # 숫자 라벨 쓰기
		csvF.write(','.join(sdata) + "\n")         # 리스트 이미지 데이터 -> 문자열 변환 후 쓰기

		
		
		#break # 성공하면 빼면 됨
```
