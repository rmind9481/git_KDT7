# [Tensor 의 Storage]
-	메모리에 원소/요소가 저장된 정보및 데이터 가지는 속성

```python
# 모듈 로딩
import numpy as np
import torch 
```

```python
# 텐서 생성
t1 = torch.tensor([[11,22,33,44,55],[66,77,88,99,00]])
```

```python
def printStorage(obj, obj_name):
	print(f'\n ======={obj_name} 기본정보 ======')
	print(f'Shape	{obj.shape}')
	print(f'Dim		{obj.ndim}D')
	print(f'Dtype	{obj.dtype}')
	print(f'itemsize{obj.itemsize}')
	print(f'======= STORAGE ======')
	print('Offset', obj.storage_offset())
	print('Stride', obj.stride())
	print(obj)
	print(f'========================')
	print(obj.untyped_storage())
```

```python
# 메모리 저장 정보 
printStorage(t1, 't1')
```

* 텐서 형태변경

```python
t2 = t1.reshape(5,-1)
printStorage(t2,'t2')
```

* 축(axis) 변경

```python
# t1 (2,5) ==> t1(5,2)
t2 = t1.transpose(0,1)
printStorage(t2,'t2')
print(t2)
```

```python
# tensor 형태 변경
# t2.view(-1,10) ## => 연속성이 깨져서 생긴 오류
print(t2.is_contiguous())

t2 =t2.contiguous() # 새롭게 생성

print(t2.is_contiguous())
t2.view(-1,10)
```
