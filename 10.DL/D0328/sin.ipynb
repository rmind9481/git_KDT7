```python

#   

import math 	# 
import torch 	#   
import matplotlib.pyplot as plt #  

# -pi  pi   1000 
x = torch.linspace(-math.pi, math.pi, 1000)
print(x)

#     y 
y = torch.sin(x)
print(y)


#     ()  y 
a = torch.randn(())
b = torch.randn(())
c = torch.randn(())
d = torch.randn(())

#    3  
y_random = a*x**3 + b* x**2 + c* x+d

#    y  
plt.subplot(2,1,1)
plt.title('Y true')
plt.plot(x,y)

#      y  
plt.subplot(2,1,2)
plt.title('y_pred')
plt.plot(x, y_random)

#   
plt.show()
```
