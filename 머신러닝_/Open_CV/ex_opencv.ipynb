# 이미지 출력(OpenCV)

```python
# cv2.imshow("Image Window", pic_arr)  # "Image Window"는 창 이름
# cv2.waitKey(0)  # 키 입력을 기다려 창이 바로 닫히지 않도록 설정
# cv2.destroyAllWindows()  # 모든 창 닫기

# 주피터 환경에서는 실행이 힘들다.
```

```python
# URL에서 이미지 다운로드
response = requests.get(url)  # URL 요청
image_np = np.array(bytearray(response.content), dtype=np.uint8)  # 바이트 데이터를 NumPy 배열로 변환
image = cv2.imdecode(image_np, cv2.IMREAD_COLOR)  # OpenCV 이미지 변환

# 이미지가 정상적으로 로드되었는지 확인
if image is None:
    print("이미지를 불러올 수 없습니다!")
else:
    # OpenCV는 BGR, Matplotlib은 RGB이므로 변환 필요
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Matplotlib을 사용하여 이미지 출력
    plt.imshow(image_rgb)
    plt.axis("off")
    plt.show()
```

-   이미지 데이터는 본래 개별 원소로 이루어진 행렬의 집합
-   여기서 개별 원소는 픽셀, 개별 원소의 값은 픽셀의 강도 0(검정)~255(흰색)

#### 주의점
OpenCV 는 기본적으로 이미지를 BGR 타입으로 읽는다. => MAtplotlib 등 대부분의 이미지 는 RGB 타입

BGR -> RBG 타입으로 변환

```python
# 컬러 이미지 로드
image_bgr = cv2.imread('./images/plane.jpg', cv2.IMREAD_COLOR)

# RBG 타입으로 변환
image_rbg = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# plot
plt.imshow(image_rbg); plt.show()
```

### 이미지 저장

- resize() 메소드를 이용 이미지크기 변경 가능
- 256x256 크기의 이미지를 로드한후 이를 50x50 크기의 이미지로 변경

```python
image = cv2.imread('./images/plane_256x256.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# 이미지 크리를 50x50 으로 변경
image_50x50 = cv2.resize(image, (50,50))
```

```python
# 출력
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original Image')
ax[1].imshow(image_50x50, cmap='gray')
ax[1].set_title('Resized Image')

plt.show()
```

### 이미지 자르기 blur 처리
-   이미지를 흐리게 하기 위해서 각 픽셀을 주변 픽셀의 평균값으로 변환
-   픽셀에 수행되는 연산을 커널
-   커널이 클수록 이미지가 더 부드러워 진다

```python
image = cv2.imread('./images/plane_256x256.jpg', cv2.IMREAD_COLOR)

# blur(): 각 픽셀에 커널 개수의 역수를 곱하여 모두 더함 
image_blurry = cv2.blur(image,(5,5))

plt.imshow(image_blurry)
plt.show()
```

```python
# 100x100 커널과 같이 큰 커널을 적용 하면

image_very_blurry = cv2.blur(image, (100,100))

plt.imshow(image_very_blurry)
plt.show()
```

### 이미지 선명하게 표현

```python
image = cv2.imread('./images/plane_256x256.jpg', cv2.IMREAD_COLOR)

# 커널 생성(대상이 있는 픽셀을 강조)
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# 커널 적용 
image_sharp = cv2.filter2D(image, -1, kernel)

fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original Image')
ax[1].imshow(image_sharp, cmap='gray')
ax[1].set_title('Sharp Image')

```

# 이미지 대비 높이기

```python
image = cv2.imread('./images/plane_256x256.jpg', cv2.IMREAD_GRAYSCALE)

# 이미지 대비를 향상
image_enhanced = cv2.equalizeHist(image)

# plot
fig, ax = plt.subplots(1,2, figsize=(10, 5))
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original Image')
ax[1].imshow(image_enhanced, cmap='gray')
ax[1].set_title('Enhanced Image')
```
